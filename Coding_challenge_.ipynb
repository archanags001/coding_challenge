# Part 1

```python
class Tensor:
    def __init__(self, data, shape):
        self.shape = shape
        self.data1 = data
        try:
          self.data = []
          for x in self.data1:
            for y in x:
              self.data.append(y)
        except:
          pass
        self.tensor = self.total_elements()

        tensor_shape = self.shape[::-1]
        # If empty list return [] else return tensor
        if self.shape == []:
            print([])
        else:
            for i in tensor_shape[:-1]:
                self.tensor = self.create_tensor(self.tensor, i)
            #print(self.tensor)

    def tensor_to_list(self):
        return self.tensor

    # Function for finding total number of elements and adjust the number of element (pad zeros to the list or trim the list)based on size
    def total_elements(self):
        elements = 1
        for i in self.shape:
            elements = elements * i

        if elements < len(self.data):
            return self.data[:elements]
        elif elements == len(self.data):
          return self.data[::]
        else:
            number_of_zeros = elements - len(self.data)

            self.data.extend([0] * number_of_zeros)
            return self.data

    # Function for convert list  to tensor based on shape
    def create_tensor(self, list_data, i):
        tensor = []
        while list_data != []:
            tensor.append(list_data[:i])
            list_data = list_data[i:]
        return tensor
```

```python
import tensorflow as tf
import numpy as np
```

```python
mnist =tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128,activation="sigmoid"),
                                  #  tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(10, activation="softmax")
                                   ])
```

```python
predictions = model(x_train[:1]).numpy()
predictions
```

```python
tf.nn.softmax(predictions).numpy()
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
loss_fn(y_train[:1], predictions).numpy()
```

```python
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test,  y_test, verbose=2)
```

```python
predict_x = model.predict(x_test)
predict_model=np.argmax(predict_x,axis=1)
```

```python
class Dense:
  def __init__(self,weights, biases):
    # self.input = input
    self.weights = weights
    self.biases = biases

  # Dot product
  def dot(self, A, B):
    dot_out = [[sum(a * b for a, b in zip(A_row, B_col)) 
                              for B_col in zip(*B)]
                              for A_row in A]
    # print("dot shape",np.shape(dot_out))
    # print("input shape: ",np.shape(A))
    # print("weight shape: ",np.shape(B))
    return dot_out

  #Sigmoid 
  def sigmoid(self,x):
    e = 2.71828
    for i in range(len(x)):
      for j in range(len(x[0])):
        x[i][j] =  1/(e **(-(x[i][j]))+1)

    return x

  #Softmax
  def softmax(self,x):
    e = 2.71828
    for_sum = []
    d = []
    for i in range(len(x)):
      for j in range(len(x[0])):
        x[i][j] = e**x[i][j]
    for i in range(len(x)):
      for_sum.append(sum(x[i]))

    for i in range(len(x)):
      for j in range(len(x[0])):
        x[i][j] = x[i][j]/for_sum[i]
    return x

  # Forward Propagation
  def forword(self,input):
    layer1 = self.sigmoid((self.dot(input,self.weights[0]))+self.biases[0])
    layer2 = self.sigmoid(self.dot(layer1,self.weights[1])+self.biases[1])
    output = self.softmax(layer2)
    result = output[0].index(max(output[0]))
    return result


  
 
```
